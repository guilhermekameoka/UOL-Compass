```python
# Data visualization and processing dependencies
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

# Data Visualization and Preprocessing

### General

```python
# Reads csv to pandas dataframe
hotel_data = pd.read_csv('hotel_reservations.csv')
hotel_data
```

```python
# Shows column names
hotel_data.columns
```

```python
# Removes Booking_ID and booking_status
hotel_data.drop(columns = ['Booking_ID', 'booking_status'], inplace = True)
hotel_data
```

```python
# Count null values
hotel_data.isnull().sum()
```

### Categorical columns (One Hot Encoder)

```python
# Count plot for better type_of_meal_plan visualization
sns.countplot(x = hotel_data['type_of_meal_plan'])
```

```python
# Count plot for better room_type_reserved visualization
sns.countplot(x = hotel_data['room_type_reserved'])
```

```python
# Count plot for better market_segment_type visualization
sns.countplot(x = hotel_data['market_segment_type'])
```

```python
# Extract and concat dummies from categorical columns: 
#   type_of_meal_plan
hotel_data = pd.concat([hotel_data, pd.get_dummies(hotel_data['type_of_meal_plan'])], axis='columns')
#   room_type_reserved
hotel_data = pd.concat([hotel_data, pd.get_dummies(hotel_data['room_type_reserved'])], axis='columns')
#   market_segment_type
hotel_data = pd.concat([hotel_data, pd.get_dummies(hotel_data['market_segment_type'])], axis='columns')
# Removes categorical columns
hotel_data.drop(columns = ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type'], inplace = True)
hotel_data.head()
```

### label_avg_price_per_room creation and visualization

```python
# Creates classifier column label
hotel_data.loc[(hotel_data['avg_price_per_room'] <= 85, 'label_avg_price_per_room')] = 1
hotel_data.loc[((hotel_data['avg_price_per_room'] > 85) & (hotel_data['avg_price_per_room'] < 115), 'label_avg_price_per_room')] = 2
hotel_data.loc[(hotel_data['avg_price_per_room'] >= 115, 'label_avg_price_per_room')] = 3
# Removes avg_price_per_room
hotel_data.drop(columns = ['avg_price_per_room'], inplace = True)
hotel_data
```

```python
# Places classification label in start of dataframe, needed for aws xgboosting
columns = [column_name for column_name in hotel_data.columns if column_name != 'label_avg_price_per_room']
columns.insert(0, 'label_avg_price_per_room')
hotel_data = hotel_data[columns]
hotel_data.columns
```

```python
# Counts how many rows has a specific label from label avg_price_per_room
np.unique(hotel_data['label_avg_price_per_room'], return_counts = True)
```

```python
# Count plot for better label_avg_price_per_room visualization
sns.countplot(x = hotel_data['label_avg_price_per_room'])
```

```python
# Heatmap for label_avg_price_per_room to visualize correlation
fig = plt.figure(figsize=(10,10))
sns.heatmap(hotel_data.corr()[['label_avg_price_per_room']], annot=True)
```

```python
# Removes columns that have low correlation, below -0.20
hotel_data = hotel_data.drop(columns = ['Room_Type 1', 'Offline'])
fig = plt.figure(figsize=(10,10))
sns.heatmap(hotel_data.corr()[['label_avg_price_per_room']], annot=True)
```

### Column normalization

```python
# Normalizes only label_avg_price_per_room for xgboost
hotel_data['label_avg_price_per_room'] = (hotel_data['label_avg_price_per_room'] - hotel_data['label_avg_price_per_room'].min()) / (hotel_data['label_avg_price_per_room'].max() - hotel_data['label_avg_price_per_room'].min())
hotel_data.head(10)
```

### Data split for test and train

```python
# Separates train and test by a limiar
train_test_limiar = int(0.70*len(hotel_data))

train = hotel_data.iloc[:train_test_limiar, :]
test = hotel_data.iloc[train_test_limiar:, :]
hotel_data.shape, train.shape, test.shape
```

```python
# separates sets in axis
x_train = train.iloc[:, 1:].values
y_train = train.iloc[:, 0].values

x_test = test.iloc[:, 1:].values
y_test = test.iloc[:, 0].values

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Saves csv files from parsed train and test to send to s3 later
train.to_csv('hotel_reservations_train_xgboost_with_header.csv', header = True, index = False)
train.to_csv('hotel_reservations_train_xgboost.csv', header = False, index = False)
test.to_csv('hotel_reservations_test_xgboost.csv', header = False, index = False)
```

# SageMaker Configuration

```python
# aws sagemaker connection
import boto3
import sagemaker
from sagemaker import Session
import sagemaker.amazon.common as smac

# data parsing to store and receive from aws
import io
import os
```

```python
# Session, role, region and Storing parameters
session = sagemaker.Session()
role = sagemaker.get_execution_role()
region_name = boto3.Session().region_name
bucket = 'uc-mlaws-sagemakerbucket'
train_data_path = 'compass-s5/train/hotel_reservations_train'
test_data_path = 'compass-s5/test/hotel_reservations_test'
model_path = 'compass-s5/models/hotel_reservations_xgboost_model'
region_name, train_data_path, test_data_path, model_path
```

```python
# Sends bynary train csv to s3
with open('hotel_reservations_train_xgboost.csv', 'rb') as file:
    boto3.Session().resource('s3').Bucket(bucket).Object(os.path.join(train_data_path)).upload_fileobj(file)
```

```python
# Sends bynary test csv to s3
with open('hotel_reservations_test_xgboost.csv', 'rb') as file:
    boto3.Session().resource('s3').Bucket(bucket).Object(os.path.join(test_data_path)).upload_fileobj(file)
```

# Training

```python
# Creates aws container for xgboost
container = sagemaker.image_uris.retrieve(framework = 'xgboost', region = region_name, version = '1.7-1')
```

```python
# Creates aws estimator for xgboost using 1 instance of ml.m5.large and stop sessions
xgboost = sagemaker.estimator.Estimator(image_uri = container,
                                       role = role,
                                       instance_count = 1,
                                       instance_type = 'ml.m5.large',
                                       output_path = 's3://{}/{}'.format(bucket, model_path),
                                       sagemaker_session = session,
                                       use_stop_session = True)
's3://{}/{}'.format(bucket, model_path)
```

```python
# Sets xgboost hyperparameters
xgboost.set_hyperparameters(num_round = 200, objective = 'reg:logistic')
```

```python
# creates xgboost data channels using s3 stored binary files
xgboost_train_set = sagemaker.inputs.TrainingInput(s3_data = 's3://{}/{}'.format(bucket, train_data_path), content_type = 'csv', s3_data_type = 'S3Prefix')
xgboost_valid_set = sagemaker.inputs.TrainingInput(s3_data = 's3://{}/{}'.format(bucket, test_data_path), content_type = 'csv', s3_data_type = 'S3Prefix')
xgboost_data_channels = { 'train': xgboost_train_set, 'validation': xgboost_valid_set }
's3://{}/{}'.format(bucket, train_data_path), 's3://{}/{}'.format(bucket, test_data_path)
```

```python
# Train model
xgboost.fit(xgboost_data_channels)
```

# Validation

```python
from sagemaker.serializers import CSVSerializer
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
# Deploy classifier endpoint
xgboost_classifier = xgboost.deploy(initial_instance_count = 1, instance_type = 'ml.m5.large')
```

```python
# Do test previsions
xgboost_classifier.serializer = CSVSerializer()
previsions = np.array(xgboost_classifier.predict(x_test).decode('utf-8').split(',')).astype(np.float32)
previsions
```

```python
# Creates class labels from previsions and y_test float values
class_previsions = [1 if (prevision <= 1/3) else (2 if (prevision <= 2/3) else 3) for prevision in previsions ]
class_y_test = [1 if (test <= 1/3) else (2 if (test <= 2/3) else 3) for test in y_test ]
class_previsions[:10], class_y_test[:10]
```

```python
# Confusion matrix
cm = confusion_matrix(class_y_test, class_previsions)
cm
```

```python
# Accuracy
accuracy_score(class_y_test, class_previsions)
```

```python
# Classification report
print(classification_report(class_y_test, class_previsions))
```
